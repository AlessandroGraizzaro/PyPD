```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
def func(x, a, b, c):
    return a * np.exp(-b*x) + c
```

```python
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
np.random.seed(1729)
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b-', label='data')
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
popt
```

```python
plt.plot(xdata, func(xdata, *popt), 'r-', label='fit: a%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
```

```python
plt.plot(xdata, func(xdata, popt[0], popt[1], popt[2]))
```

```python
coefs = np.polyfit(x, y, 10)
ffit = np.polyval(coefs, x)
```

```python
new_x = np.linspace(0,26)
new_ffit = np.polyval(coefs, new_x)
```

```python
plt.plot(x, y, 'o', label='Raw')
plt.plot(x, ffit, 'x', label='Fit to Raw')
plt.plot(new_x, new_ffit, label='Fit to Linspace')
plt.legend()
plt.grid(True)
```
